```python


import cv2
import numpy as np
import tensorflow as tf

# Load the pre-trained face detection model from OpenCV
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the pre-trained gender classification model using TensorFlow
gender_model = tf.keras.models.load_model('model.h5')  # Replace with the actual path to your model

print(gender_model)

# Function to detect and classify gender
def detect_and_classify_face(frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    for (x, y, w, h) in faces:
        # Draw a rectangle around the face
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        
        # Extract the face region and preprocess for gender classification
        face_roi = gray[y:y+h, x:x+w]
        face = cv2.resize(face_roi, (110, 80))
        face = face / 255.0  # Normalize the pixel values
        face = np.expand_dims(face, axis=-1)  # Add channel dimension
        
        # Classify the gender
        gender_prediction = gender_model.predict(np.array([face]))
        
        # Get the gender label
        gender_label = "Male" if 0.0 <= gender_prediction[0][0] < 0.5 else "Female"

        
        # Display the gender label
        cv2.putText(frame, gender_label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
    
    return frame


# Access the camera
cap = cv2.VideoCapture(0)  # Use 0 for the default camera, or specify a different camera index

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Detect and classify faces
    frame = detect_and_classify_face(frame)
    
    # Display the frame with detected faces and gender labels
    cv2.imshow('Face Detection and Gender Classification', frame)
    
    # Break the loop if the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close the OpenCV window
cap.release()
cv2.destroyAllWindows()
```
